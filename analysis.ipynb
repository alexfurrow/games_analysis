```python
#Now make it extensible
def retention_calc(df,period):
    retained = df[
    (df['auth_ts'] != df['reg_ts']) &
    ((df["auth_ts"] - df["reg_ts"]) <= period) & 
    ((df["auth_ts"] - df["reg_ts"]) >= 0)
    ]
    days    = period // 86400
    rem1    = period %  86400
    hours   = rem1  // 3600
    rem2    = rem1  %  3600
    minutes = rem2  // 60
    seconds = rem2  %  60

    print(f'{retained.shape[0]} users were retained over {days} days, {hours} hours, and {minutes} minutes')

retention_calc(retention_df,11186400)
```
